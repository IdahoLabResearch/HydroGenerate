```python
flow = pd.read_csv('data_test.csv') # pandas data frame
flow['dateTime'] = pd.to_datetime(flow['dateTime']) # preprocessing convert to datetime
flow = flow.set_index('dateTime') # set datetime index # flolw is in cfs
```

```python
# Test with daily data to make it faster
d_flow = flow.groupby(pd.Grouper(freq='D')).mean()           # compute the daily mean and drop first and last row as these can be incomplete
d_flow.head()
flow = d_flow.copy()
```
