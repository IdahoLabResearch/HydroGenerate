## Turbine selection based on design flow and head

In our previous version, HydroGenerate would select the turbine type based solely on head. The new feature shown below, is an improvement in the turbine selection process.

Based on head value and design flow, the turbine will be selected using the map shown below. The point (design_flow, head) will be used to determine where in the regions for each turbine the site falls under. For cases where the point falls under two overlapping regions, HydroGenerate will chose the one with the shortest distance to the centroid of the area. The options will be listed in `hp.turbine_type_dict`, where hp is the object returned by `calculate_hp_potential`.

```python
# Using flow as a pandas dataframe and adding annual energy calculation

# Note: When using a pandas dataframe as flow data, please set the datetime index before 
# using HydroGenerate.  (https://pandas.pydata.org/docs/reference/api/pandas.DatetimeIndex.html)

head = 20 # ft
power = None
penstock_length = 50 # ft
hp_type = 'Diversion' 

hp = calculate_hp_potential(flow= flow, rated_power= power, head= head,
                            pctime_runfull = 30,
                            penstock_headloss_calculation= True,
                            design_flow= None,
                            electricity_sell_price = 0.05,
                            resource_category= 'CanalConduit',
                            hydropower_type= hp_type, penstock_length= penstock_length,
                            flow_column= 'discharge_cfs', annual_caclulation= True)

# hp.dataframe_output[hp.dataframe_output.turbine_flow_cfs.eq(0)]
```

```python
# Command to diplay the flow vs head turbine selection map. Each area corresponds to a turbine type and the brown dot denotes the combination of head and flow for a given site. 
hp.turbinetype_figure
```

## Flow Constraints

This feature allows the user to specify flow constraints that denote annual and major maintenance (5 years):
1. Annual maintenance: the flow will be set to zero for **one week** in the month with the lowest average flow every year. Turn on setting the parameter `annual_maintenance_flag` to **True**
2. Major maintenance: the flow will be set to zero for **two weeks** in the month with lowest average flow every 5 years. Turn on setting the parameter `major_maintenance_flag` to **True**

The user can specify the threshold of flow in cubic-feet-sec using the paramentes `minimum_turbineflow`, or in percent of design flow `minimum_turbineflow_percent`.

```python
hp_maintenance_ex = calculate_hp_potential(flow= flow, flow_column='discharge_cfs', rated_power= power, head= head, penstock_headloss_calculation= True,
                                           hydropower_type= hp_type, penstock_length= penstock_length, 
                                           minimum_turbineflow = 5000, 
                                           # minimum_turbineflow_percent = 80,
                                           annual_caclulation= True, 
                                           annual_maintenance_flag= True, major_maintenance_flag= True)
```

```python
# Show the data where flow is zero
hp_maintenance_ex.dataframe_output[hp_maintenance_ex.dataframe_output.turbine_flow_cfs.eq(0)]
```

```python
# High-level statistics of the dataframe
hp_maintenance_ex.dataframe_output.describe()
```

```python
df = hp_maintenance_ex.dataframe_output.copy()

#fig = plt.figure(figsize=(8, 6))
fig2, ax1 = plt.subplots()

color_plot = 'tab:red'
ax1.set_xlabel('Days')
ax1.set_ylabel('Flow rate (cfs)', color=color_plot)
ax1.plot(df['turbine_flow_cfs'], label="Flow rate", color=color_plot)
ax1.tick_params(axis='y', labelcolor=color_plot)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color_plot2 = 'tab:blue'
ax2.set_ylabel('Power (kW)', color=color_plot2)  # we already handled the x-label with ax1
ax2.plot(df['power_kW'],'b-',label="Power", color=color_plot2)
ax2.tick_params(axis='y', labelcolor=color_plot2)
ax1.grid(True, axis='both', color='k',linestyle='--',alpha=0.4)
plt.title("Yearly flow data and potential power")
fig2.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
